```python
from bs4 import BeautifulSoup

import requests

import re

import pandas as pd

base_url = "https://www.fantacalcio.it"

headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML,)'
}


r = requests.get("https://www.fantacalcio.it/squadre")


soup = BeautifulSoup(r.content, 'lxml')


team_list = soup.find_all("h3", class_="inline-block pull-left")

team_link = [] 

for item in team_list:
    for link in item.find_all("a", href=True):
        team_link.append(link["href"])
    
test_link = "https://www.fantacalcio.it/squadre/Atalanta"


team_summary = []

for link in team_link:


    r = requests.get(link, headers=headers)


    soup = BeautifulSoup(r.content, 'lxml')


    pattern = "(?<=>).*(?=<)"


    num_pattern = "\d"


    name = re.findall(pattern, str(soup.find_all("h1")))[0].strip()

    manager = soup.find_all("span")[6].text.strip()

    founded = soup.find_all("span")[7].text.strip()

    website = soup.find_all("span")[8].text.strip()

    stadium_capacity = "".join(re.findall(num_pattern, soup.find_all("li", class_="col-lg-6 col-md-6 col-sm-12 col-xs-12")[3].text.strip()))



    team = {
        "Team": name,
        
        "Manager": manager,
        
        "Founded": founded,
        
        "Website": website,
        
        "Stadium_capacity": stadium_capacity
        
    }
    
    team_summary.append(team)


df = pd.DataFrame(team_summary)


df.to_csv("team_summary.csv", index=False)
```
